```python
# Importing the libraries
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
#Loading the dataset
df = pd.read_csv('AviationData.csv', encoding='cp1252', low_memory= False)
df.head()
```

```python
#Checking Data types and missing values 
df.info()
```

```python
#Replacing '.'with '_' for readability
df.columns = df.columns.str.replace('.','_')
df.columns
```

```python
### Dropping Columns we wont be needing for analysis and they also have lots of missing values


df=df.drop(['Schedule', 'Air_carrier' ,'FAR_Description', 'Longitude', 'Latitude', 'Airport_Code', 'Airport_Name'], axis=1)
df.head()
```

```python
#Changing Data types of Event_date to Date 
df['Event_Date'] = pd.to_datetime(df['Event_Date'])
df['Event_Date']
```

```python
#Checking the null values
df.isnull().sum().sort_values(ascending=False)
```
